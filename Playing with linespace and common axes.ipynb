<h1> Matplotlib basics </h1>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,2*(np.pi),101) #linear space numpy.
#linspace(start, stop, num = 50, endpoint = True, retstep = False, dtype = None) 
#: Returns number spaces evenly w.r.t interval.

y=np.sin(x)

xnum=np.linspace(0,7,15)  #(start,end,steps)
ynum=np.linspace(-1,1,11)

plt.plot(x,y,color='r',label='sin')
plt.xlabel('Angle in radians')
plt.ylabel('Amplitude')
plt.title('Plot of trignometric functions')
plt.legend()
plt.xticks(xnum)  #no.of points on x axis
plt.yticks(ynum)  #no. of points on y axis
plt.show()
```

```python
x=np.linspace(0,2*(np.pi),101) #linear space numpy.
#linspace(start, stop, num = 50, endpoint = True, retstep = False, dtype = None) 
#: Returns number spaces evenly w.r.t interval.

y=np.sin(x)
z=np.cos(x)

xnum=np.linspace(0,7,15)  #(start,end,steps)
ynum=np.linspace(-1,1,11)

plt.plot(x,y,'r',x,z,'g')
plt.xlabel('Angle in radians')
plt.ylabel('Amplitude')
plt.title('Plot of trignometric functions\n')
#plt.legend()
plt.xticks(xnum)  #no.of points on x axis
plt.yticks(ynum)  #no. of points on y axis
plt.show()
```

<h3>Making twinx box fancier </h3>

```python

x=np.linspace(0,2*(np.pi),101) #linear space numpy.
#linspace(start, stop, num = 50, endpoint = True, retstep = False, dtype = None) 
#: Returns number spaces evenly w.r.t interval.

y=np.sin(x)
z=np.sinh(x)

xnum=np.linspace(0,7,15)  #(start,end,steps)
ynum1=np.linspace(-1,1,11)
ynum2=np.linspace(0,300,7)

#plt.plot(x,y,'r',x,z,'g')
fig,ax1=plt.subplots()  #axes has seperate figure object and axes object
ax2=ax1.twinx() #ax2 will have common x axis as ax1.
curve1,= ax1.plot(x,y, color='g',label='sinx')
curve2,= ax2.plot(x,z,color='b',label='sinhx')
curves=[curve1,curve2]

#ax1.legend()  #only one legend shows up
#ax2.legend()
ax1.legend(curves,[curve.get_label() for curve in curves])  #for all legends
#ax2.legend(curves,[curve.get_label() for curve in curves]) any one would work

ax1.set_xlabel('angle/value',color=curve1.get_color())
#ax2.set_xlabel('angle/value',color=curve1.get_color())  does not work

ax1.set_ylabel('Amplitude',color=curve1.get_color())
ax2.set_ylabel('Amplitude',color=curve2.get_color())

ax1.tick_params(axis='x', colors=curve1.get_color())
ax1.tick_params(axis='y', colors=curve1.get_color())
ax2.tick_params(axis='y', colors=curve2.get_color())

ax1.set_yticks(ynum1) #setting y axis for green line
ax2.set_yticks(ynum2) #setting y axis for blue line

ax1.set_xticks(xnum) #setting x axis for both lines as they have common x axis

ax1.grid(color=curve1.get_color())
#ax2.grid(color=curve2.get_color())

plt.title('Plot of trignometric functions')
 
plt.show()
```

<h1> Plots with common y-axis and different x-axis </h1>

```python
import numpy as np
import matplotlib.pyplot as plt

y = np.linspace(0, 2.0*np.pi, 101)
x1 = np.sin(y)
x2 = np.sinh(y)

# values for making ticks in x and y axis
ynum = np.linspace(0, 7, 15)
xnum1 = np.linspace(-1, 1, 11)
xnum2 = np.linspace(0, 300, 7)

fig, ax1 = plt.subplots()

ax2 = ax1.twiny() # ax2 and ax1 will have common y axis and different x axis

# plot the curves on axes 1, and 2, and get the axes handles
curve1, = ax1.plot(x1, y, label="sin", color='g')
curve2, = ax2.plot(x2, y, label="sinh", color='b')

# Make a curves list to access the parameters in the curves
curves = [curve1, curve2]


# ax1.legend() # will not display the legend of ax2
# ax2.legend() # will not display the legend of ax1
# ax1.legend(curves, [curve.get_label() for curve in curves])
ax2.legend(curves, [curve.get_label() for curve in curves])

# x axis labels via the axes
ax1.set_xlabel("Amplitude", color=curve1.get_color())
ax2.set_xlabel("Amplitude", color=curve2.get_color())

# y axis label via the axes
ax1.set_ylabel("Angle/Value", color=curve1.get_color())


# y ticks - make them coloured as well
ax1.tick_params(axis='y', colors=curve1.get_color())


# x axis ticks via the axes
ax1.tick_params(axis='x', colors=curve1.get_color())
ax2.tick_params(axis='x', colors=curve2.get_color())

# set x ticks
ax1.set_xticks(xnum1)
ax2.set_xticks(xnum2)

# set y ticks
ax1.set_yticks(ynum)
# ax2.set_yticks(ynumbers) # also works


# To make grids using axes 2
ax1.grid(color=curve2.get_color())
ax2.grid(color=curve2.get_color())
ax1.xaxis.grid(False)


plt.title("Plot of sine and hyperbolic sine \n \n")
plt.show()
```
